# **Modelando M√°quinas de Vetores de Suporte (SVM)**

O modelo de *Support Vector Machines* (SVM) ou M√°quina de Vetores de Suporte (SVM) √© um algoritmo importante no arsenal do aprendizado de m√°quina, se destacam por sua capacidade de lidar com dados complexos e realizar classifica√ß√µes precisas.

O SVM tem √≥timo desempenho em problemas de classifica√ß√£o, especialmente em conjuntos de dados n√£o lineares de pequeno a m√©dio porte. Ele funciona encontrando um hiperplano que separa os dados em diferentes categorias, mesmo quando os dados n√£o s√£o linearmente separ√°veis.

## **Como funciona o SVM?**

1. **Mapeamento de Dados para Espa√ßos de Alta Dimens√£o**

Em muitos casos, os dados originais n√£o podem ser separados linearmente em classes distintas. Para lidar com essa limita√ß√£o, o SVM transforma os dados para um espa√ßo de alta dimens√£o (espa√ßo de caracter√≠sticas) onde eles possam ser mais facilmente separados. Por exemplo, imagine pontos distribu√≠dos em uma linha sem separa√ß√£o clara entre classes. Ao projetar esses pontos em um espa√ßo bidimensional, utilizando coordenadas como ùë• e ùë•2, as classes podem se tornar linearmente separ√°veis. Essa transforma√ß√£o permite ao SVM trabalhar em problemas complexos onde os dados originalmente n√£o apresentam separabilidade linear.

2. **Encontrando o Hiperplano √ìtimo**

Ap√≥s mapear os dados, o SVM busca o hiperplano que melhor separa as classes. Os dados devem ser transformados de forma que um separador possa ser desenhado como um hiperplano, em que os pontos de dados se enquadram em categorias diferentes, representando um conjunto de dados n√£o separ√°vel linearmente.

Desta forma √© poss√≠vel transferir esses dados para um espa√ßo de dimens√£o superior, por exemplo, mapeando-os para um espa√ßo tridimensional. Ap√≥s a transforma√ß√£o, o limite entre as duas categorias pode ser definido por um hiperplano. Num espa√ßo tridimensional, o separador √© mostrado como um plano. Este plano pode ser usado para classificar casos novos ou desconhecidos. Portanto, o algoritmo SVM produz um hiperplano ideal que categoriza novos exemplos.

Em outras palavras, se os dados est√£o dispostos em uma linha no espa√ßo 1D e n√£o s√£o separ√°veis, podem ser projetados para um espa√ßo 2D (como e ) onde se tornam separ√°veis. Ap√≥s o mapeamento, o SVM  identifica o hiperplano √≥timo que divide os dados em
classes distintas. Em um espa√ßo 2D, esse hiperplano √© uma linha, enquanto em um espa√ßo 3D, √© um plano.
Em dimens√µes mais altas, √© chamado de hiperplano.

3. **Maximizando a Margem**

O SVM n√£o busca apenas um hiperplano que separa as classes, mas sim aquele que maximiza a margem entre as duas categorias. A margem √© definida como a dist√¢ncia entre o hiperplano e os pontos mais pr√≥ximos de cada classe. Esses pontos s√£o chamados de **vetores de suporte**, pois determinam a posi√ß√£o e a orienta√ß√£o do hiperplano.

Esse processo de maximiza√ß√£o da margem confere ao SVM uma maior capacidade de generaliza√ß√£o, minimizando erros ao classificar novos exemplos.

4.  **Kerneliza√ß√£o**

Como nem sempre os dados podem ser separados linearmente, mesmo em espa√ßos de alta dimens√£o. Para resolver isso, o SVM utiliza uma t√©cnica chamada kerneliza√ß√£o, que aplica fun√ß√µes matem√°ticas (kernels) para transformar os dados em um novo espa√ßo onde a separa√ß√£o linear seja poss√≠vel, s√£o eles:

* **Linear**: Mant√©m os dados no espa√ßo original.
* **Polinomial**: Projeta os dados em uma dimens√£o polinomialmente maior.
* **RBF (Radial Basis Function)**: Projeta os dados em um espa√ßo infinito com base em dist√¢ncias relativas.
* **Sigmoide**: Projeta os dados de forma n√£o linear, similar a redes neurais.

Em suma, a kerneliza√ß√£o permite que o SVM seja aplicado em dados n√£o linearmente separ√°veis, tornando-o extremamente flex√≠vel e eficaz em uma ampla variedade de problemas.

Brevemente, resumo suas aplica√ß√µes e limita√ß√µes, vantagens e desvantagens, entre outros aspectos.

**Aplica√ß√µes:**

SVMs s√£o eficazes em uma variedade de tarefas, incluindo:

* **Classifica√ß√£o:**  Desde a categoriza√ß√£o de imagens at√© a detec√ß√£o de spam, as SVM demonstram excelente performance em problemas de classifica√ß√£o, especialmente quando a separa√ß√£o entre classes n√£o √© linear.
* **Regress√£o:**  Embora menos comum, as SVM tamb√©m podem ser aplicadas em problemas de regress√£o, prevendo valores cont√≠nuos.
* **Detec√ß√£o de Novidades:**  SVM podem identificar padr√µes incomuns ou *outliers* em conjuntos de dados.

**Performance e Escalabilidade**

O SVM √© particularmente eficaze em conjuntos de dados de pequeno a m√©dio porte (centenas a milhares de inst√¢ncias) com rela√ß√µes n√£o lineares entre as vari√°veis, mas seu desempenho pode ser afetado em conjuntos de dados muito grandes, devido √† complexidade computacional. O SVM pode apresentar desafios de escalabilidade.  Nesses casos, algoritmos alternativos como √°rvores de decis√£o ou redes neurais podem ser mais eficientes.

**Vantagens:**

* Eficazes em espa√ßos de alta dimens√£o.
* Robustas a *outliers*.
* Vers√°teis, com diferentes fun√ß√µes kernel para modelar rela√ß√µes complexas.

**Desvantagens:**

* **Computacionalmente caros para grandes conjuntos de dados**: a complexidade treinamento de um modelo SVM envolve a otimiza√ß√£o de um problema quadr√°tico, que se torna computacionalmente caro √† medida que o tamanho do conjunto de dados aumenta. O tempo de treinamento pode crescer significativamente com o n√∫mero de amostras, tornando os SVM poucos eficientes para conjuntos de dados massivos. Embora o tempo de predi√ß√£o seja geralmente menor que o tempo de treinamento, este pode ser afetado pelo tamanho do conjunto de dados e pelo n√∫mero de vetores de suporte, consumindo muita mem√≥ria e impactando o desempenho.
* **Escolha da fun√ß√£o kernel e dos hiperpar√¢metros pode se mostrar complexa**: o modelo utiliza fun√ß√µes kernel para mapear os dados em um espa√ßo de alta dimens√£o. Existem diversos tipos de kernels (linear, polinomial, RBF, sigmoide), cada um com caracter√≠sticas e par√¢metros pr√≥prios.
  * A escolha do kernel ideal depende do problema e dos dados, o que pode exigir experimenta√ß√£o e conhecimento especializado. Al√©m disso, o SVM possui outros hiperpar√¢metros, como o par√¢metro de regulariza√ß√£o (C) e a toler√¢ncia para o crit√©rio de parada.Encontrar a combina√ß√£o ideal de hiperpar√¢metros para um bom desempenho do modelo pode ser desafiador e demandar t√©cnicas de otimiza√ß√£o e valida√ß√£o cruzada.
  * A interpreta√ß√£o do impacto de cada hiperpar√¢metro no modelo pode ser complexa, exigindo conhecimento te√≥rico e pr√°tico sobre SVMs.
* Caso se pretenda uma modelagem preditiva, o SVM n√£o fornece diretamente estimativas de probabilidade.

A coluna BareNuc inclui alguns valores que n√£o s√£o num√©ricos, √© preciso remover essas linhas.

```python
# Converter os valores da coluna 'BareNuc' para num√©ricos, substituindo entradas inv√°lidas por NaN
cell_df = cell_df[pd.to_numeric(cell_df['BareNuc'], errors='coerce').notnull()]

# Remover as linhas que cont√™m NaN na coluna 'BareNuc'
cell_df = cell_df.dropna(subset=['BareNuc'])

# Alterar o tipo de dados da coluna 'BareNuc' para inteiro
cell_df['BareNuc'] = cell_df['BareNuc'].astype('int')

# Verificar os tipos de dados das colunas do DataFrame
cell_df.dtypes
```

```python
feature_df = cell_df[['Clump', 'UnifSize', 'UnifShape', 'MargAdh', 'SingEpiSize', 'BareNuc', 'BlandChrom', 'NormNucl', 'Mit']]
X = np.asarray(feature_df)
X[0:5]
```

```python
# Definir a lista de colunas de interesse
interestCols = ['Clump', 'UnifSize', 'UnifShape', 'MargAdh', 'SingEpiSize',
                     'BareNuc', 'BlandChrom', 'NormNucl', 'Mit']

# Verificar se todas as colunas de interesse est√£o presentes no DataFrame
for coluna in interestCols:
    if coluna not in cell_df.columns:
        raise ValueError(f"A coluna '{coluna}' n√£o est√° presente no DataFrame.")

# Selecionar as colunas de interesse
feature_df = cell_df[interestCols]

# Converter o DataFrame selecionado em um array NumPy
X = feature_df.to_numpy()

# Exibir as primeiras 5 linhas do array NumPy resultante
print("As primeiras 5 linhas do array NumPy s√£o:")
print(X[:5])
```

```python
# Verificar se a coluna 'Class' existe no DataFrame
if 'Class' not in cell_df.columns:
    raise ValueError("A coluna 'Class' n√£o est√° presente no DataFrame.")

# Converter a coluna 'Class' para um array NumPy
y = cell_df['Class'].to_numpy()

# Exibir as primeiras 5 entradas do array resultante
print("As primeiras 5 entradas do array 'y' s√£o:", y[:5])
```

## **Treinando e testando o modelo**

---

### **Divis√£o dos dados: fun√ß√£o `train_test_split`**

A fun√ß√£o `train_test_split` do m√≥dulo `model_selection` do scikit-learn ser√° utilizada para dividir os arrays (matrizes) em subconjuntos de treinamento e teste de forma aleat√≥ria.

**Par√¢metros:**
* `X`: matriz de caracter√≠sticas (features).
* `y`: vetor de r√≥tulos ou vari√°veis alvo.
* `test_size=0.2`: Define que 20% dos dados ser√£o utilizados para o conjunto de teste.
* `random_state=4`: garante a reprodutibilidade da divis√£o dos dados, permitindo que os mesmos subconjuntos sejam obtidos em execu√ß√µes subsequentes.
* `stratify=y`: assegura que a divis√£o mantenha a mesma propor√ß√£o de classes em ambos os conjuntos, o que √© especialmente importante em problemas de classifica√ß√£o para evitar vieses.

```python
# Dividir os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=4, stratify=y
)

# Exibir as dimens√µes dos conjuntos resultantes
print('Conjunto de treinamento:', X_train.shape, y_train.shape)
print('Conjunto de teste:', X_test.shape, y_test.shape)
```

## **Modelagem em SVM com kernel Radial Basis Function (RBF)**

---
A modelagem em SVM envolve o seguinte processo abstrato:

1. **Transforma√ß√£o inicial**: os dados s√£o projetados para um espa√ßo de alta dimens√£o (manualmente ou via kernel).
2. **Defini√ß√£o do hiperplano √≥timo**: um separador √© encontrado para dividir as classes.
3. **Maximiza√ß√£o da margem**: a margem entre os vetores de suporte √© maximizada, garantindo uma separa√ß√£o ideal.
4. **Classifica√ß√£o**: o modelo final utiliza o hiperplano para prever a classe de novos exemplos.

Como explicitado acima, o algoritmo Support Vector Machine utiliza o conceito de "*kernelling*" para mapear dados em um espa√ßo dimensional mais alto, permitindo a classifica√ß√£o eficiente mesmo quando os dados n√£o s√£o linearmente separ√°veis. Essa transforma√ß√£o √© realizada por meio de fun√ß√µes kernel, cada uma com suas pr√≥prias caracter√≠sticas e desempenho.

Existem diversas fun√ß√µes kernel dispon√≠veis, como:

* **Linear**: eficaz para dados linearmente separ√°veis.
* **Polinomial:** mapeia os dados em um espa√ßo polinomial, √∫til para dados com rela√ß√µes n√£o lineares.
* ***Radial Basis Function*** ou Fun√ß√£o de Base Radial (RBF): uma fun√ß√£o popular que mapeia os dados em um espa√ßo de dimens√£o infinita, geralmente oferecendo bom desempenho em diversos cen√°rios.
* **Sigmoide**: similar √† fun√ß√£o sigmoide utilizada em redes neurais.

A escolha da fun√ß√£o kernel ideal depende do conjunto de dados e do problema em quest√£o. Uma vez que n√£o existe uma regra universal para determinar a melhor op√ß√£o, a experimenta√ß√£o com diferentes fun√ß√µes e a compara√ß√£o dos resultados s√£o cruciais para encontrar a fun√ß√£o mais adequada.

```python
# Acessar o classificador SVM
from sklearn import svm
# Padroniza os dados
from sklearn.preprocessing import StandardScaler
# Cria um pipeline para facilitar o processamento sequencial
from sklearn.pipeline import make_pipeline

# Supondo que X_train e y_train j√° estejam definidos

# Criar um pipeline que inclui a padroniza√ß√£o dos dados e o classificador SVM com kernel RBF
clf = make_pipeline(
    StandardScaler(),  # Padroniza os dados para ter m√©dia 0 e vari√¢ncia 1
    svm.SVC(kernel='rbf', C=1.0, gamma='scale')  # Inicializa o classificador SVM
)

# Treinar o modelo com os dados de treinamento
clf.fit(X_train, y_train)
```

```python
# Realizar previs√µes no conjunto de teste
yhat = clf.predict(X_test)

# Exibir as primeiras 5 previs√µes
print("As primeiras 5 previs√µes s√£o:", yhat[:5])
```

### ***Trade-off* entre Precision e Recall**

A curva Precision-Recall fornece uma vis√£o mais detalhada do desempenho em rela√ß√£o √† classifica√ß√£o bin√°ria, especialmente quando as classes est√£o desbalanceadas. O *trade-off* entre *Precision* e *Recall* reflete a necessidade de ajustar o desempenho do modelo √†s prioridades do problema em quest√£o. O gr√°fico abaixo fornece algumas informa√ß√µes para se encontrar o equil√≠brio adequado, considerando-se as implica√ß√µes dos erros de classifica√ß√£o no dom√≠nio da aplica√ß√£o.

Precision e Recall rabalham em oposi√ß√£o. Ao se maximizar a precis√£o, tende-se a sacrificar a abrang√™ncia (recall), pois o modelo torna-se mais conservador nas previs√µes. E vice-versa, ao se maximizar o recall, a precis√£o pode cair, pois o modelo tenta classificar mais amostras como positivas. O *threshold* determina o ponto de corte usado para classificar uma amostra como positiva ou negativa:
* **Limiar baixo** (ex.: pr√≥ximo de 0, aumenta o Recall): mais amostras ser√£o classificadas como positivas, incluindo muitas da classe verdadeira, o que tamb√©m aumenta o n√∫mero de Falsos Positivos, reduzindo a Precision.
* **Limiar alto** (ex.: pr√≥ximo de 1, aumenta a Precision): apenas as amostras com alta confian√ßa s√£o classificadas como positivas, mas algumas amostras positivas podem ser perdidas (Falsos Negativos), reduzindo o Recall.



```python
from sklearn.metrics import precision_recall_curve

# Obter pontua√ß√µes da fun√ß√£o de decis√£o
y_scores = clf.decision_function(X_test)

# precision_recall_curve utiliza r√≥tulos bin√°rios (0/1 or -1/1),
# Especificar pos_label para indicar qual classe √© considerada "positiva" para o c√°lculo da curva.
precisions, recalls, thresholds = precision_recall_curve(y_test, y_scores, pos_label=4)

plt.plot(thresholds, precisions[:-1], "b--", label="Precision", linewidth=2)
plt.plot(thresholds, recalls[:-1], "g-", label="Recall", linewidth=2)
threshold = 0.6  # Exemplo de valor limite
plt.vlines(threshold, 0, 1.0, "k", "dotted", label="threshold")

plt.xlabel("Threshold")
plt.ylabel("Precision/Recall")
plt.title("Precision-Recall Curve")
plt.legend()
plt.grid(True)
plt.show()
```

## **Visualiza√ß√£o Kernel Linear versus RBF**

* **Linear**: uma linha reta separa as duas classes, para dados linearmente separ√°veis.
* **RBF**: separa√ß√£o n√£o linear ajusta-se √†s curvas formadas pelos dados, adequado para dados mais complexos.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_blobs

# 1. Gerar dados sint√©ticos
X, y = make_blobs(n_samples=100, centers=2, random_state=42, cluster_std=1.5)

# 2. Treinar o modelo SVM com kernel linear e RBF
svm_linear = SVC(kernel="linear", C=1)
svm_rbf = SVC(kernel="rbf", C=1, gamma=0.5)

svm_linear.fit(X, y)
svm_rbf.fit(X, y)

# 3. Fun√ß√£o para visualiza√ß√£o do plano de decis√£o
def plot_decision_boundary(svm, X, y, ax, title):
    # Criar grade de pontos para o gr√°fico
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                         np.linspace(y_min, y_max, 200))
    # Calcular valores da fun√ß√£o de decis√£o
    Z = svm.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plotar os dados e o plano de decis√£o
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, edgecolors="k", s=50)
    ax.contour(xx, yy, Z, levels=[-1, 0, 1], linestyles=["--", "-", "--"], colors="k")
    ax.scatter(svm.support_vectors_[:, 0], svm.support_vectors_[:, 1], s=200,
               facecolors="none", edgecolors="red", label="Vetores de Suporte")
    ax.set_title(title)
    ax.set_xlabel("Feature 1")
    ax.set_ylabel("Feature 2")
    ax.legend()

# 4. Criar gr√°ficos lado a lado
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

# Kernel Linear
plot_decision_boundary(svm_linear, X, y, axes[0], title="SVM com Kernel Linear")

# Kernel RBF
plot_decision_boundary(svm_rbf, X, y, axes[1], title="SVM com Kernel RBF")

plt.tight_layout()
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.datasets import load_iris

# 1. Carregar o dataset Iris
iris = load_iris()
X = iris["data"][:, (2, 3)]  # Usar comprimento e largura das p√©talas
y = (iris["target"] == 2).astype(int)  # Iris virginica como classe positiva

# 2. Ajustar dois modelos SVM com diferentes valores de C
svm_clf_low_c = SVC(kernel="linear", C=1)
svm_clf_high_c = SVC(kernel="linear", C=100)
svm_clf_low_c.fit(X, y)
svm_clf_high_c.fit(X, y)

# 3. Fun√ß√£o para calcular margens e suportes
def plot_svm_decision_boundary(svm_clf, X, y, ax, title):
    # Plotar os dados
    ax.scatter(X[:, 0][y == 1], X[:, 1][y == 1], color="green", s=60, marker="^", label="Iris virginica")
    ax.scatter(X[:, 0][y == 0], X[:, 1][y == 0], color="blue", s=60, marker="s", label="Iris versicolor")

    # Gerar grade para decis√£o
    x0s = np.linspace(X[:, 0].min(), X[:, 0].max(), 200)
    x1s = np.linspace(X[:, 1].min(), X[:, 1].max(), 200)
    x0, x1 = np.meshgrid(x0s, x1s)
    X_new = np.c_[x0.ravel(), x1.ravel()]
    y_pred = svm_clf.decision_function(X_new).reshape(x0.shape)

    # Plotar fronteiras de decis√£o
    ax.contour(x0, x1, y_pred, colors="black", levels=[-1, 0, 1], alpha=0.7, linestyles=["--", "-", "--"])

    # Destacar vetores de suporte
    ax.scatter(
        svm_clf.support_vectors_[:, 0],
        svm_clf.support_vectors_[:, 1],
        s=200, facecolors="none", edgecolors="gray", label="Vetores de Suporte"
    )
    ax.set_title(title)
    ax.set_xlabel("Petal length")
    ax.set_ylabel("Petal width")
    ax.legend()

# 4. Criar gr√°ficos lado a lado
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

plot_svm_decision_boundary(svm_clf_low_c, X, y, axes[0], title="C = 1")
plot_svm_decision_boundary(svm_clf_high_c, X, y, axes[1], title="C = 100")

plt.tight_layout()
plt.show()
```

O gr√°fico acima compara claramente o impacto do par√¢metro C, efeito da regulariza√ß√£o em aprendizado de m√°quina, em um classificador SVM.

* **C baixo** (margens amplas): Maior flexibilidade, maior n√∫mero de viola√ß√µes.
* **C alto** (margens estreitas): Classifica√ß√£o mais r√≠gida, menor n√∫mero de viola√ß√µes.
